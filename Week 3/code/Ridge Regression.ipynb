```python
import numpy as np

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset = np.loadtxt('blogData_train.csv', delimiter=",")
```

```python
dataset.shape
```

```python
X = dataset[:,0:-1]
y = dataset[:,-1]
```

```python
n_alphas = 200
alphas = np.logspace(-10, -2, n_alphas)
clf = linear_model.RidgeCV(alphas=alphas)
clf.fit(X, y)
```

```python
clf.alpha_
```

```python
clf.intercept_
```

```python
clf.coef_
```

```python
testds = np.loadtxt('blogData_test-2012.03.31.01_00.csv', delimiter=',')
```

```python
Xte = testds[:,0:-1]
yte = testds[:,-1]
```

```python
clf.score(Xte, yte)
```

```python
mean_squared_error(yte, clf.predict(Xte))
```

```python
with open('../code/params.txt', 'w') as f:
    for c in clf.coef_:
        f.write("{}\n".format(c))
    
    f.write("{}\n".format(clf.intercept_))
```
